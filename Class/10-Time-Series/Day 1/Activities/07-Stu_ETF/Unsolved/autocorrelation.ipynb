```python
import pandas as pd
%matplotlib inline
from pathlib import Path
csvpath = Path('../Resources/high_frequency_euro_ETF_bid_ask_spreads.csv')
euro_df = pd.read_csv(csvpath, parse_dates=True, index_col='timestamp')
euro_df.head()
```

```python
# Resample to 10 Seconds ('10S'), take the mean, and drop the NaN values
# YOUR CODE HERE!
seconds_df = euro_df.resample("10S").mean().dropna()
seconds_df.head()
```

```python
# Plot the resulting bid-ask spread data
euro_df.plot()
```

```python
# Calculate the autocorrelation
# YOUR CODE HERE!
euro_df.bid_ask_spread.autocorr(lag=1)
```
